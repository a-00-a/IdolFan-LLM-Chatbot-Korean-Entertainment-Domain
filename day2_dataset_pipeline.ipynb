```python
# 데이터셋 로딩
from datasets import load_dataset

dataset = load_dataset("imdb")
print(dataset)
```

```python
print(dataset["train"][0])
```

```python
# 정제함수
def clean_text(example):
    text = example["text"]
    text = text.replace("<br />", " ")
    text = text.strip()
    return {"text": text}

dataset = dataset.map(clean_text)
```

```python
# 토그나이저 로딩
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token # Add this line to set a padding token

# 토크나이징 (split별 안전 처리)
def tokenize(example):
    return tokenizer(example["text"],
                     truncation=True,
                     padding="max_length",
                     max_length=128)

tokenized_train = dataset["train"].map(tokenize, batched=True)
tokenized_test = dataset["test"].map(tokenize, batched=True)
```

```python
# 결과 확인
print(tokenized_train[0].keys())
print(tokenized_test[0].keys())
```
